```python
import numpy as np
arr = np.array([10, 20, 30, 40, 50])

print("Array:")
print(arr)
print("\nData Type:")
print(type(arr))
print("\nShape:")
print(arr.shape)
print("\nDimensions:")
print(arr.ndim)
print("\nSize:")
print(arr.size)
print("\nAddition (+10):")
print(arr + 10)
print("\nMultiplication (*2):")
print(arr * 2)
print("\nDivision (/2):")
print(arr / 2)
print("\nZeros Array:")
print(np.zeros(5))
print("\nOnes Array:")
print(np.ones(5))
print("\nRange Array (1 to 10):")
print(np.arange(1, 11))
arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])
print("\n2D Array:")
print(arr2)
print("\n2D Array Shape:")
print(arr2.shape)
print("\n2D Array Dimensions:")
print(arr2.ndim)
```
